## Introduction

In my field, bioinformatics, data is messy because life processes are messy. There's often no way to prove that your algorithm is the correct one, only the one that minimizes fitting error with respect to the data. To address this, you can use **synthetic data**, which provides logically consistent and statistically tunable ground truth data.

![Here to Help](misc/here_to_help.png)

Currently, there are fantastic resources for processing data in Python, but no centralized resource for generating data. 
* Tied to a particular research question
* Not well-maintained
* In MATLAB

I started **Synthpy**, a library for the generation of synthetic data using the Python scientific stack, to experiment with the idea that many parts of generating data could sit behind an API, allowing you to focus on engineering algorithms and testing their assumptions.

The way it works is that **synthetic objects**, resembling components of raw data, are equipped with **hooks** where the user can insert their own behaviour. Some basic pre-defined behaviours are provided (i.e. brownian motion), but it's largely left up to you what you want your synthetic objects to do.

### Example 1:  Quantifying interactions between DNA repair enzymes using stochastic modeling

Some people from my lab were interested in whether two DNA repair enzymes, <span style="color:red">pDNA-PKcs</span> and <span style="color:green">LigIV</span>, were significantly co-localized or were randomly distributed. It wasn't good enough to calculate how much overlap happened between the green and red channels because random dispersal would result in some degree of overlap, so how do you know to what degree the overlap is un-random?

Their solution was to calculate the "interaction factor" of an image using the ratio of the amount of overlaps expected by random versus the amount of overlap observed. This actually allows the measurement of interaction to be density-independent and independent of the size and shape of the objects simulated.

$$\frac{1}{n} \sum^{n}_{k=1} \frac{P_{k}(0)}{1 - (1 - P_{k}(0)) * IF}$$

```python
# first, pre-process the image using scikit-image tools
import numpy as np
from skimage.io import imread, imshow
from skimage import measure
from skimage.filters import gaussian, threshold_otsu
from scipy.ndimage import find_objects
%matplotlib inline

def segment_channel(cell_img, ch, sigma=10):
    """Custom function for image segmentation based on scikit-image
    API. Since Synthpy doesn't do processing, this is separate."""
    channel_actors = np.zeros(cell_img.shape, dtype=np.uint8)
    channel_actors[:, :, ch] = cell_img[:, :, ch]
    
    blur = gaussian(channel_actors, sigma=sigma, multichannel=True)
    bval = threshold_otsu(blur.sum(axis=2))
    blobs = measure.label(blur > bval, background=0)
    blob_slices = find_objects(blobs)
    
    img_slices = []
    for b_slice in blob_slices:
        channel_slice = channel_actors[b_slice].sum(axis=2)
        img_slice = np.zeros(channel_slice.shape + (3,), dtype=np.uint8)
        img_slice[:, :, ch] = channel_slice
        img_slices.append(img_slice)
    return img_slices
```

```python
# show how the interaction looks with random localization
from synthpy.image import (FileImage, SynthImage, SynthImageStack,
                           brownian, random_loc, uniform_noise, even_grid, 
                           random_loc_adjust, random_index_adjust)

# load images and preprocess (not shown)
cell_img = imread('misc/clusters.tif')
red_segs = segment_channel(cell_img, 0, sigma=10)
green_segs = segment_channel(cell_img, 1, sigma=10)

# automatically make instances of actors
red = FileImage.generate(20, red_segs, 'foo')
green = FileImage.generate(20, green_segs, 'bar')

# attach hooks to actors as they are added to the stage
stage = SynthImage()
stage.add_actors(green, random_loc)
stage.add_actors(red, random_loc)
stage.show()
```

```python
def interaction_factor(actor, field, threshold=0.7):
    """This is an example of a function you can feed into
    the synthetic object hooks. Locations are randomly generated,
    but if a random number isn't higher than the threshold,
    the location is re-generated until there is an overlap."""
    y_high, x_high = field.shape[:2]
    y_loc = np.random.randint(0, y_high)
    x_loc = np.random.randint(0, x_high)
    if np.random.random() <= threshold:
        while not field.img[y_loc, x_loc, 1]:
            y_loc = np.random.randint(0, y_high)
            x_loc = np.random.randint(0, x_high)
    actor.loc = (y_loc, x_loc)
    return

# add custom behaviour to a set of actors
stage = SynthImage()
stage.add_actors(green, random_loc)
stage.add_actors(red, interaction_factor)
stage.show()
```

### Example 2: Synthetic time series of cell migration

Wound healing assays are popular for studying cell migration. Basically, you lay down a layer of cells and scratch a "wound" through them using a needle. Then, you take photos using a microscope at set time intervals. 

Cell migration is a combination of:
* Forward motion to fill in the wound
* Brownian motion that randomly alters the path
* Cyclical motion of a cell pushing itself forward.

```python
# create a scratch mask
scratch = np.zeros((512, 512))
border = 300
scratch[:, :border] = 1
imshow(scratch)

# define wound healing behaviour
# this is dependent on the distance from the scratch *and* time
def wound_healing(actor, stage, border=512, delta=2, forward=20):
    """Tells the cells to use brownian motion except for a forward
    force that is a function of the edge distance and a cyclical
    force that is a function of time."""
    y, x = actor.loc
    t = actor.index
    br_y = y + np.random.randn() * 2*delta
    br_x = x + np.random.randn() * 2*delta
    ff = 1 - (border-x)/border
    ff_x = x + (np.sin(t)+1) * forward
    n_y = int(br_y)
    n_x = int((1-ff)*br_x + ff*ff_x)
    actor.loc = (n_y, n_x)
    return
```

```python
img = imread('misc/nuclei.tif')
nuclei = segment_channel(img, 2, sigma=1)
blue = FileImage.generate(45, nuclei, 'dapi', time_loc_func=wound_healing)

stage = SynthImage()
stage.add_actors(blue, even_grid, {'mask': scratch})

# these are one-off adjustments where the function is not saved to the state
stage.adjust_actors(random_loc_adjust)
stage.adjust_actors(random_index_adjust)
stage.show()
```

```python
# when you build the stack, the functions fed to hooks auto-generate the behaviour
stack = SynthImageStack(stage, n_stages=20)
stack.build_stack()

# cells migrate forward as a function of their closeness to the edge of the wound
imshow(stack.stack[4])
```
